### k-NN Classifier Class

```python
# creating the kNN class
class initial_KNN:
    def __init__(self, k=3, encoder_type='resnets', distance_metric='euclidean'):
        # initialising the hyperparameters
        self.k = k
        self.distance_metric = distance_metric
        self.encoder_type = encoder_type
    
    def fit(self, X, Y):
        self.X_train = X[:, 0]
        if self.encoder_type=='resnets':
            self.X_train = X[:, 0]
        elif self.encoder_type=='vits':
            self.X_train = X[:, 1]
#         print(self.X_train.shape)
        self.Y_train = Y
        
    def calculate_distance(self, x1, x2):
        # calculating the distance based on distance metric
        if self.distance_metric == 'euclidean':
            return np.sqrt(np.sum((x1-x2)**2))
        if(self.distance_metric == 'manhattan'):
            return np.sum(np.abs(x1-x2))
        if(self.distance_metric == 'cosine'):
            distance = dot(x1, np.transpose(x2))/(norm(x1) * norm(x2))
#             print(distance[0][0])
            return 1-distance[0][0]
        
    def _predict(self, x):
        # predicting the class based on k nearest neighbours
        distances = [self.calculate_distance(x, x_train) for x_train in self.X_train]
        indices = sorted(range(len(distances)), key=lambda i: distances[i])
#         indices = np.argsort(distances)
        k_nearest_indices = indices[:self.k]
        k_nearest_labels = [self.Y_train[i] for i in k_nearest_indices]
        most_common = Counter(k_nearest_labels).most_common()
        
        # if their is a tie in case of most common nearest label 
        if len(most_common)>1:
            weighted_votes = {}
            for i, label in zip(k_nearest_indices, k_nearest_labels):
                if label in weighted_votes:
                    weighted_votes[label] += 1.0 / (1.0 + distances[i])  # Weighted vote based on distance
                else:
                    weighted_votes[label] = 1.0 / (1.0 + distances[i])

            # Finding the label with the highest weighted vote
            predicted_label = max(weighted_votes, key=weighted_votes.get)

            return predicted_label
        return most_common[0][0]
    
    def predict(self, X):
        predictions = [self._predict(x) for x in X]
        return predictions
    
    def validate(self, X, Y_actual):
        # returning the validation metrics
        Y_predictions = self.predict(X)
        f1_weighted = f1_score(Y_actual, Y_predictions, average='weighted', zero_division=1)
        f1_micro = f1_score(Y_actual, Y_predictions, average='micro', zero_division=1)
        f1_macro = f1_score(Y_actual, Y_predictions, average='macro', zero_division=1)
        accuracy = accuracy_score(Y_actual, Y_predictions)
        precision = precision_score(Y_actual, Y_predictions, average='weighted', zero_division=1)
        recall = recall_score(Y_actual, Y_predictions, average='weighted', zero_division=1)
        return recall, precision, accuracy, f1_weighted, f1_micro, f1_macro
```

### Optimising the class for kNN classifier using vectorization

```python
# most optimised version of kNN class
class optimised_KNN:
    def __init__(self, k=3, encoder_type='resnets', distance_metric='euclidean'):
        self.k = k
        self.distance_metric = distance_metric
        self.encoder_type = encoder_type
    
    def fit(self, X, Y):
        self.X_train = X[:, 0]
        if self.encoder_type=='resnets':
            self.X_train = X[:, 0]
        elif self.encoder_type=='vits':
            self.X_train = X[:, 1]
#         print(self.X_train.shape)
        self.Y_train = Y
        
    def calculate_distance(self, x1, x2):
        if self.distance_metric == 'euclidean':
            return np.sqrt(np.sum((x1-x2)**2))
        if(self.distance_metric == 'manhattan'):
            return np.sum(np.abs(x1-x2))
        if(self.distance_metric == 'cosine'):
            distance = dot(x1, np.transpose(x2))/(norm(x1) * norm(x2))
#             print(distance[0][0])
            return 1-distance[0][0]
        
    def _predict(self, x):
        distances = [self.calculate_distance(x, x_train) for x_train in self.X_train]
        indices = np.argsort(distances)
        k_nearest_indices = indices[:self.k]
        k_nearest_labels = [self.Y_train[i] for i in k_nearest_indices]
        most_common = Counter(k_nearest_labels).most_common()
        
        if len(most_common)>1:
            weighted_votes = {}
            for i, label in zip(k_nearest_indices, k_nearest_labels):
                if label in weighted_votes:
                    weighted_votes[label] += 1.0 / (1.0 + distances[i])  # Weighted vote based on distance
                else:
                    weighted_votes[label] = 1.0 / (1.0 + distances[i])

            # Finding the label with the highest weighted vote
            predicted_label = max(weighted_votes, key=weighted_votes.get)

            return predicted_label
        return most_common[0][0]
    
    def predict(self, X):
        predictions = [self._predict(x) for x in X]
        return predictions
    
    def validate(self, X, Y_actual):
        Y_predictions = self.predict(X)
        f1_weighted = f1_score(Y_actual, Y_predictions, average='weighted', zero_division=1)
        f1_micro = f1_score(Y_actual, Y_predictions, average='micro', zero_division=1)
        f1_macro = f1_score(Y_actual, Y_predictions, average='macro', zero_division=1)
        accuracy = accuracy_score(Y_actual, Y_predictions)
        precision = precision_score(Y_actual, Y_predictions, average='weighted', zero_division=1)
        recall = recall_score(Y_actual, Y_predictions, average='weighted', zero_division=1)
        return recall, precision, accuracy, f1_weighted, f1_micro, f1_macro
```

```python
def hamming_score(y_true: np.ndarray, y_pred: np.ndarray):
        numerator = (y_true & y_pred).sum(axis=1)
        denominator = (y_true | y_pred).sum(axis=1)

        return np.divide(numerator, denominator, out=np.ones_like(numerator, dtype=np.float_),
                         where=denominator != 0).mean()
```

### Decision Tree Classifier based on Powerset Formulation

```python
# define the class for decision tree classifier using powerset formulations
class PowersetDecisionTree(BaseEstimator, ClassifierMixin):
    def __init__(self, max_depth=None, max_features=None, criterion='gini'):
        self.max_depth = max_depth
        self.max_features = max_features
        self.criterion = criterion
        self.clf = LabelPowerset(DecisionTreeClassifier(
            max_depth=self.max_depth,
            max_features=self.max_features,
            criterion=self.criterion,
            random_state = 42
        ))

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        y_encoded = self.clf.predict(X)
        return y_encoded.toarray()
    
    def validate(self, X, Y_actual):
        Y_predictions = self.predict(X)
        f1_micro = round(f1_score(Y_actual, Y_predictions, average='micro', zero_division=1), 4)
        f1_macro = round(f1_score(Y_actual, Y_predictions, average='macro', zero_division=1), 4)
        accuracy = round(accuracy_score(Y_actual, Y_predictions), 4)
        precision = round(precision_score(Y_actual, Y_predictions, average='weighted', zero_division=0), 4)
        recall = round(recall_score(Y_actual, Y_predictions, average='weighted', zero_division=0), 4)
        hamming_l = round(hamming_loss(Y_actual, Y_predictions), 4)
        confusion_matrix = multilabel_confusion_matrix(Y_actual, Y_predictions)
        
        return recall, precision, accuracy, f1_micro, f1_macro, hamming_l, confusion_matrix
```

```python
# Instantiating the PowersetDecisionTree
powerset_tree = PowersetDecisionTree(max_depth=5, max_features=None, criterion='gini')

# Fitting the model to your training data
powerset_tree.fit(X_train, y_train)

# Making predictions on your validation data
y_pred = powerset_tree.predict(X_test)
print(y_pred.shape)
recall, precision, accuracy, f1_micro, f1_macro, hamming_l, confusion_matrix = powerset_tree.validate(X_test, y_test)
print("For Decision Tree Classifier with Powerset Formulation")
print("recall: ", recall)
print("precision: ", precision)
print("accuracy: ", accuracy)
print("f1 score micro: ", f1_micro)
print("f1 score macro: ", f1_macro)
print("hamming loss: ", hamming_l)
print("Confusion_matrix: \n", confusion_matrix)
```

### Decision Tree Classifier based on MultiOutput Formulation

```python
# define the class for decision tree classifier using multioutput formulation
class MultiOutputDecisionTree(BaseEstimator, ClassifierMixin):
    def __init__(self, max_depth=None, max_features=None, criterion='gini'):
        self.max_depth = max_depth
        self.max_features = max_features
        self.criterion = criterion
        self.clf = clf = ClassifierChain(DecisionTreeClassifier(
            max_depth=self.max_depth,
            max_features=self.max_features,
            criterion=self.criterion, 
            random_state = 1
        ))

    def fit(self, X, y):
        self.clf.fit(X, y)

    def predict(self, X):
        y_encoded = self.clf.predict(X)
        return y_encoded.toarray()
    
    def validate(self, X, Y_actual):
        Y_predictions = self.predict(X)
        f1_micro = round(f1_score(Y_actual, Y_predictions, average='micro', zero_division=1), 4)
        f1_macro = round(f1_score(Y_actual, Y_predictions, average='macro', zero_division=1), 4)
        accuracy = round(accuracy_score(Y_actual, Y_predictions), 4)
        precision = round(precision_score(Y_actual, Y_predictions, average='weighted', zero_division=0), 4)
        recall = round(recall_score(Y_actual, Y_predictions, average='weighted', zero_division=0), 4)
#         Y_predictions = Y_predictions.astype(np.int64)
#         Y_actual = Y_actual.astype(np.int64)
#         print(type(Y_predictions))
        confusion_matrix = multilabel_confusion_matrix(Y_actual, Y_predictions)
        
        hamming_l = round(hamming_loss(Y_actual, Y_predictions), 4)
        return recall, precision, accuracy, f1_micro, f1_macro, hamming_l, confusion_matrix
```

### Reporting the k-fold validation metrics 

```python
# reporting the k-fold validation metrics for the best hyperparameters for powerset Decision tree
best_max_depth = 5
best_max_features = 11
best_criterion = 'entropy'
```

```python
k_fold = KFold(n_splits = 7, shuffle = True, random_state= 1)
f1_micro_scores = []
f1_macro_scores = []
accuracy_scores = []
precision_scores = []
recall_scores = []
hamming_losses = []
```

```python
# performing k-fold splitting
for train_indices, val_indices in k_fold.split(encoded_features):
    X_train = encoded_features[train_indices]
    X_val = encoded_features[val_indices]
    y_train = encoded_labels[train_indices]
    y_val = encoded_labels[val_indices]
    
    powerset_model = PowersetDecisionTree(max_depth = best_max_depth, max_features = best_max_features, criterion = best_criterion)
    powerset_model.fit(X_train, y_train)
    
    y_pred = powerset_model.predict(X_val)
    
    f1_micro_fold = f1_score(y_val, y_pred, average='micro')
    f1_macro_fold = f1_score(y_val, y_pred, average='macro')
    accuracy_fold = accuracy_score(y_val, y_pred)
    precision_fold = precision_score(y_val, y_pred, average='weighted')
    recall_fold = recall_score(y_val, y_pred, average='weighted')
    hamming_loss_fold = hamming_loss(y_val, y_pred)
    
    f1_micro_scores.append(f1_micro_fold)
    f1_macro_scores.append(f1_macro_fold)
    accuracy_scores.append(accuracy_fold)
    precision_scores.append(precision_fold)
    recall_scores.append(recall_fold)
    hamming_losses.append(hamming_loss_fold)
```

```python
# finding the average of all validation metrics
f1_micro_avg = np.mean(f1_micro_scores)
f1_macro_avg = np.mean(f1_macro_scores)
accuracy_avg = np.mean(accuracy_scores)
precision_avg = np.mean(precision_scores)
recall_avg = np.mean(recall_scores)
hamming_loss_avg = np.mean(hamming_losses)
```

```python
print("f1 score(micro): ", f1_micro_avg)
print("f1 score(macro): ", f1_macro_avg)
print("accuracy: ", accuracy)
print("precision: ", precision)
print("recall: ", recall)
print("hamming loss: ", hamming_loss_avg)
```
